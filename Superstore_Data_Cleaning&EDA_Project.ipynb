Data Cleaning Process

```python
#Reading of csv file
df = pd.read_csv(r"C:\Users\GBENGA\Downloads\Superstore data\ORDERS.csv", sep=',', engine='c', encoding='utf-8', encoding_errors='ignore')
```

```python
#Duplicating raw dataset 
df_copy = df.copy()
df_copy
```

```python
#Exploring the dataset on each column
df_copy.info()
```

```python
#Removing of unwanted column
df_copy.drop('DATASET', axis=1, inplace=True)
df_copy
```

```python
#Changing all the date columns data type from object to date data type.
df_copy['ORDERDATE']= pd.to_datetime(df_copy['ORDERDATE'])
df_copy['SHIPDATE']= pd.to_datetime(df_copy['SHIPDATE'])
```

```python
#Replacing the NaN value with 0.0 value
df_copy['PRODUCTBASEMARGIN']= df_copy['PRODUCTBASEMARGIN'].fillna(value=0.0)
```

```python
#Removing of duplicate
df_copy.drop_duplicates()
```

```python
#Removing all white space within the product name
df_copy['PRODUCTNAME']= df_copy['PRODUCTNAME'].str.strip()
```

```python
#Xhecking each column info after replacing NaN
df_copy.info()
```
